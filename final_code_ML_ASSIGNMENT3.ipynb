```python
# ========= Compact RNN TS Lab (fast) =========
!pip -q install tensorflow==2.15.0
import os, random, numpy as np, pandas as pd, matplotlib.pyplot as plt, tensorflow as tf
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras import layers, regularizers, callbacks, Model, Input, optimizers

# ---- Speed & Repro ----
SEED=42; random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)
plt.rcParams["figure.figsize"]=(9,3.5); tf.get_logger().setLevel('ERROR')

# ---- Metrics ----
def rmse(y_true,y_pred): return float(np.sqrt(mean_squared_error(y_true,y_pred)))
def mae(y_true,y_pred):  return float(mean_absolute_error(y_true,y_pred))
def mape(y_true,y_pred):
    y_true, y_pred = np.asarray(y_true), np.asarray(y_pred)
    return float(np.mean(np.abs((y_true-y_pred)/np.maximum(np.abs(y_true),1e-8)))*100)

# ---- Transforms / Scale / Windows ----
class StationarityTransform:
    def __init__(self,strategy='identity',seasonal_period=None):
        self.strategy=strategy; self.s=seasonal_period
    def fit_transform(self,y:pd.Series):
        y=y.astype(float)
        if self.strategy=='identity': return y.dropna()
        if self.strategy=='diff':     return y.diff().dropna()
        if self.strategy=='log':      return np.log(y.replace(0,np.nan)).dropna()
        if self.strategy=='log1p':    return np.log1p(y).dropna()
        if self.strategy=='logdiff':  return np.log(y.replace(0,np.nan)).diff().dropna()
        if self.strategy=='seasonal_diff': return y.diff(self.s or 1).dropna()
        raise ValueError("bad strategy")

class SeriesScaler:
    def __init__(self,kind="standard"):
        self.scaler = StandardScaler() if kind=="standard" else MinMaxScaler()
    def fit_transform(self,y:pd.Series):
        return pd.Series(self.scaler.fit_transform(y.values.reshape(-1,1)).ravel(), index=y.index)

def impute_series(y,method="interp_time"):
    if method=="ffill": return y.ffill().bfill()
    if method=="interp_time": return y.interpolate(method="time").ffill().bfill()
    if method=="median": return y.fillna(y.median())
    return y

def make_supervised(y:pd.Series, lookback=30, horizon=1, extra_features:pd.DataFrame=None):
    yv=y.values.astype(np.float32); X=[]; Y=[]
    extra = None if extra_features is None else extra_features.values.astype(np.float32)
    for t in range(lookback,len(yv)-horizon+1):
        x=yv[t-lookback:t].reshape(lookback,1)
        if extra is not None: x=np.concatenate([x, extra[t-lookback:t]],axis=1)
        X.append(x); Y.append(yv[t:t+horizon])
    return np.stack(X), np.stack(Y).squeeze(-1)

def expanding_window_splits(n, n_folds=2, min_train=365, valid_size=150, step=None):
    step=step or valid_size; k=0; tr_end=min_train
    while tr_end+valid_size<=n and k<n_folds:
        yield np.arange(tr_end), np.arange(tr_end,tr_end+valid_size); tr_end+=step; k+=1

# ---- Models ----
def make_optimizer(name, lr):
    name=name.lower()
    if name=="adam":    return optimizers.Adam(lr)
    if name=="rmsprop": return optimizers.RMSprop(lr)
    if name=="nadam":   return optimizers.Nadam(lr)
    if name=="sgdm":    return optimizers.SGD(lr, momentum=0.9, nesterov=True)
    if name=="adamw":
        try: return optimizers.AdamW(lr, weight_decay=1e-4)
        except: return optimizers.Adam(lr)
    raise ValueError("opt")

def build_elman(input_shape, units=64, l2=1e-4, dropout=0.2, rdrop=0.1, lr=1e-3, opt_name="adam"):
    x=Input(shape=input_shape)
    h=layers.SimpleRNN(units,activation="tanh",dropout=dropout,recurrent_dropout=rdrop,
                       kernel_regularizer=regularizers.l2(l2))(x)
    y=layers.Dense(1)(h)
    m=Model(x,y); m.compile(optimizer=make_optimizer(opt_name,lr), loss="mae"); return m

def build_jordan(input_shape, **k):  # same cell; input includes y_{t-1} channel pre-built in X
    return build_elman(input_shape, **k)

def build_multi_recurrent(input_shape, units1=64, units2=32, l2=1e-4, dropout=0.2, rdrop=0.1, lr=1e-3, opt_name="adam"):
    x=Input(shape=input_shape)
    h1=layers.SimpleRNN(units1,activation="tanh",return_sequences=True,dropout=dropout,
                        recurrent_dropout=rdrop,kernel_regularizer=regularizers.l2(l2))(x)
    h2=layers.SimpleRNN(units2,activation="tanh",dropout=dropout,recurrent_dropout=rdrop,
                        kernel_regularizer=regularizers.l2(l2))(h1)
    y=layers.Dense(1)(h2)
    m=Model(x,y); m.compile(optimizer=make_optimizer(opt_name,lr), loss="mae"); return m

# ---- CV training ----
def train_cv(X,Y, build_fn, cv_cfg, train_cfg, verbose=0):
    es=callbacks.EarlyStopping(patience=train_cfg.get("patience",4), restore_best_weights=True, monitor="val_loss")
    rows=[]
    for k,(tr,va) in enumerate(expanding_window_splits(len(Y), **cv_cfg),1):
        m=build_fn(input_shape=X.shape[1:])
        h=m.fit(X[tr],Y[tr], validation_data=(X[va],Y[va]), epochs=train_cfg.get("epochs",30),
                batch_size=train_cfg.get("batch_size",64), verbose=verbose, callbacks=[es])
        pred=m.predict(X[va],verbose=0).ravel()
        rows.append(dict(fold=k, MAE=mae(Y[va],pred), RMSE=rmse(Y[va],pred), MAPE=mape(Y[va],pred),
                         best_epoch=np.argmin(h.history["val_loss"])+1))
    df=pd.DataFrame(rows); mean=df[["MAE","RMSE","MAPE","best_epoch"]].mean().to_dict()
    return df, mean

# ---- Search (small, random) ----
from itertools import product
def param_grid(space):
    keys=list(space.keys()); combos=list(product(*[space[k] for k in keys])); return keys, combos
def run_search(X,Y, model_name, space, cv_cfg, train_cfg, max_trials=6, seed=SEED):
    keys, combos = param_grid(space)
    rng=np.random.default_rng(seed); rng.shuffle(combos); combos=combos[:max_trials]
    rows=[]
    def build(hp,input_shape):
        if model_name=="elman": return build_elman(input_shape, units=hp["units"], l2=hp["l2"], dropout=hp["dropout"],
                                                   rdrop=hp["rdrop"], lr=hp["lr"], opt_name=hp["optimizer"])
        if model_name=="jordan":return build_jordan(input_shape, units=hp["units"], l2=hp["l2"], dropout=hp["dropout"],
                                                   rdrop=hp["rdrop"], lr=hp["lr"], opt_name=hp["optimizer"])
        if model_name=="multi": return build_multi_recurrent(input_shape, units1=hp["units1"], units2=hp["units2"],
                                                             l2=hp["l2"], dropout=hp["dropout"], rdrop=hp["rdrop"],
                                                             lr=hp["lr"], opt_name=hp["optimizer"])
    for tup in combos:
        hp=dict(zip(keys,tup))
        df,mean=train_cv(X,Y, lambda input_shape: build(hp,input_shape), cv_cfg, train_cfg, verbose=0)
        rec={"model":model_name, **hp, **{f"cv_{k}":v for k,v in mean.items()}}
        rows.append(rec)
        print(f"[{model_name}] {hp} -> MAE={mean['MAE']:.4f} RMSE={mean['RMSE']:.4f} MAPE={mean['MAPE']:.2f}")
    return pd.DataFrame(rows).sort_values("cv_MAE")

# ---- Data prep (same datasets, quick) ----
def prep_climate_daily():
    t=pd.read_csv("DailyDelhiClimateTrain.csv",parse_dates=["date"]).set_index("date").sort_index()
    y=impute_series(t["meantemp"].astype(float),"interp_time")
    y=StationarityTransform("seasonal_diff",365).fit_transform(y)
    return SeriesScaler("standard").fit_transform(y)

def prep_air_quality_daily():
    aq=pd.read_csv("Air_quality.csv"); aq["From Date"]=pd.to_datetime(aq["From Date"],errors="coerce")
    aq=aq.set_index("From Date").sort_index()
    y=impute_series(pd.to_numeric(aq["PM2.5 (ug/m3)"],errors="coerce").resample("D").mean(),"interp_time")
    y=StationarityTransform("identity").fit_transform(y)
    return SeriesScaler("standard").fit_transform(y)

def prep_airpassengers_monthly():
    ap=pd.read_csv("AirPassengers.csv",parse_dates=["Month"]).set_index("Month").sort_index()
    y=StationarityTransform("logdiff").fit_transform(ap["#Passengers"].astype(float))
    return SeriesScaler("standard").fit_transform(y)

def prep_covid_daily():
    c=pd.read_csv("covid_19_indonesia_time_series_all.csv")
    c["Date"]=pd.to_datetime(c["Date"],errors="coerce"); c=c.set_index("Date").sort_index()
    y=impute_series(pd.to_numeric(c["New Cases"],errors="coerce").resample("D").sum(),"interp_time")
    y=StationarityTransform("log1p").fit_transform(y)
    return SeriesScaler("standard").fit_transform(y)

def prep_nifty_returns():
    s=pd.read_csv("stock.csv",parse_dates=["Date"]).set_index("Date").sort_index()
    y=np.log(pd.to_numeric(s["Close"],errors="coerce").dropna()).diff().dropna()
    y=StationarityTransform("identity").fit_transform(y)
    return SeriesScaler("standard").fit_transform(y)

def make_XY_for(y, lookback):
    X,Y=make_supervised(y,lookback=lookback,horizon=1)
    lag=pd.DataFrame({"y_lag1":pd.Series(y).shift(1)}).fillna(0.0).iloc[-len(y):]
    Xj,Yj=make_supervised(pd.Series(y),lookback=lookback,horizon=1,extra_features=lag)
    return X,Y,Xj,Yj

# ---- Small search spaces (still diverse) ----
OPTIMS=["adam","rmsprop","nadam","sgdm"]  # can add "adamw" later
LR=[5e-4,1e-3]                            # slim
DROP=[0.0,0.2]; RDROP=[0.0]; L2=[0.0,1e-4]

space_elman={"optimizer":OPTIMS,"lr":LR,"units":[32,64],"dropout":DROP,"rdrop":RDROP,"l2":L2}
space_jordan=space_elman.copy()
space_multi={"optimizer":OPTIMS,"lr":LR,"units1":[48,64],"units2":[24,32],"dropout":DROP,"rdrop":RDROP,"l2":L2}

# ---- CV/training budgets (fast) ----
cv_daily=dict(n_folds=2, min_train=365, valid_size=150, step=150)
cv_month=dict(n_folds=2, min_train=48,  valid_size=18,  step=18)
train_cfg=dict(epochs=30, batch_size=64, patience=4)
MAX_TRIALS=6  # per model per dataset

# ---- Plot helpers ----
def plot_leaderboard(df,title):
    top=df.groupby("model",as_index=False)["cv_MAE"].min().sort_values("cv_MAE")
    plt.figure(); plt.bar(top["model"], top["cv_MAE"]); plt.title(f"{title}: best CV MAE"); plt.ylabel("MAE"); plt.show()

def plot_opt_ablation(df,title):
    plt.figure()
    for m in sorted(df["model"].unique()):
        means=df[df["model"]==m].groupby("optimizer")["cv_MAE"].mean().sort_values()
        plt.plot(range(len(means)), means.values, marker="o", label=m)
    plt.xticks(range(len(means)), means.index, rotation=25); plt.title(f"{title}: optimiser ablation")
    plt.ylabel("CV MAE"); plt.legend(); plt.tight_layout(); plt.show()

def plot_lr(df,title):
    plt.figure()
    for m in sorted(df["model"].unique()):
        means=df[df["model"]==m].groupby("lr")["cv_MAE"].mean().sort_values()
        plt.plot(means.index, means.values, marker="o", label=m)
    plt.xscale("log"); plt.xlabel("LR"); plt.ylabel("CV MAE"); plt.title(f"{title}: LR sensitivity"); plt.legend(); plt.show()

def plot_capacity(df,title):
    plt.figure()
    for m in sorted(df["model"].unique()):
        key="units1" if m=="multi" else "units"
        means=df[df["model"]==m].groupby(key)["cv_MAE"].mean().sort_values()
        plt.plot(means.index, means.values, marker="o", label=m)
    plt.xlabel("units (or units1)"); plt.ylabel("CV MAE"); plt.title(f"{title}: capacity vs error"); plt.legend(); plt.show()

# ---- Single dataset runner ----
def run_dataset(name, y, lookback, cv_cfg):
    print(f"\n=== {name} ===")
    X,Y,Xj,Yj = make_XY_for(y, lookback)
    res_e = run_search(X,Y,"elman", space_elman,  cv_cfg, train_cfg, MAX_TRIALS)
    res_j = run_search(Xj,Yj,"jordan",space_jordan,cv_cfg, train_cfg, MAX_TRIALS)
    res_m = run_search(X,Y,"multi", space_multi,  cv_cfg, train_cfg, MAX_TRIALS)
    df=pd.concat([res_e,res_j,res_m]).sort_values("cv_MAE").reset_index(drop=True)
    print(df.head(5));
    plot_leaderboard(df,name); plot_opt_ablation(df,name); plot_lr(df,name); plot_capacity(df,name)
    return df

# ========= Run (adjust as time allows) =========
y_climate = prep_climate_daily()
y_aq      = prep_air_quality_daily()
y_ap      = prep_airpassengers_monthly()
y_covid   = prep_covid_daily()
y_nifty   = prep_nifty_returns()

res_climate = run_dataset("Climate (Delhi)", y_climate, lookback=30, cv_cfg=cv_daily)
res_aq      = run_dataset("Air Quality (Tirupati)", y_aq, lookback=30, cv_cfg=cv_daily)
res_ap      = run_dataset("AirPassengers", y_ap, lookback=12, cv_cfg=cv_month)
res_covid   = run_dataset("COVID-19 Indonesia", y_covid, lookback=21, cv_cfg=cv_daily)
res_nifty   = run_dataset("NIFTY-50 ZEETELE (returns)", y_nifty, lookback=30, cv_cfg=cv_daily)

summary=(pd.concat([
    res_climate.assign(dataset="climate"),
    res_aq.assign(dataset="air_quality"),
    res_ap.assign(dataset="airpassengers"),
    res_covid.assign(dataset="covid"),
    res_nifty.assign(dataset="nifty"),
]).reset_index(drop=True))
display(summary.head(15))
summary.to_csv("hparam_search_summary_fast.csv",index=False)
print("Saved: hparam_search_summary_fast.csv")
```

```python
# === Final hold-out evaluation for each model on each dataset ===
import pandas as pd, numpy as np

# 1) Use the summary from your search (change filename if needed)
summary = pd.read_csv("hparam_search_summary_fast.csv")

# 2) Pick the best config per (dataset, model) by lowest CV MAE
best_pm = summary.loc[summary.groupby(["dataset","model"])["cv_MAE"].idxmin()].reset_index(drop=True)

# 3) Datasets, lookbacks, and test sizes (align with your CV valid_size)
DATASETS = {
    "climate":       dict(series=lambda: prep_climate_daily(),       lookback=30, test_len=150),
    "air_quality":   dict(series=lambda: prep_air_quality_daily(),   lookback=30, test_len=150),
    "airpassengers": dict(series=lambda: prep_airpassengers_monthly(),lookback=12, test_len=18),
    "covid":         dict(series=lambda: prep_covid_daily(),         lookback=21, test_len=150),
    "nifty":         dict(series=lambda: prep_nifty_returns(),       lookback=30, test_len=150),
}

def build_from_row(row, input_shape):
    opt, lr, l2 = row["optimizer"], float(row["lr"]), float(row["l2"])
    drop, rdrop  = float(row["dropout"]), float(row.get("rdrop",0.0))
    if row["model"]=="elman":
        return build_elman(input_shape, units=int(row["units"]), l2=l2, dropout=drop, rdrop=rdrop, lr=lr, opt_name=opt)
    if row["model"]=="jordan":
        return build_jordan(input_shape, units=int(row["units"]), l2=l2, dropout=drop, rdrop=rdrop, lr=lr, opt_name=opt)
    if row["model"]=="multi":
        return build_multi_recurrent(input_shape, units1=int(row["units1"]), units2=int(row["units2"]),
                                     l2=l2, dropout=drop, rdrop=rdrop, lr=lr, opt_name=opt)
    raise ValueError("unknown model")

def make_XY_for(y, lookback):
    X,Y = make_supervised(y, lookback=lookback, horizon=1)
    lag = pd.DataFrame({"y_lag1": pd.Series(y).shift(1)}).fillna(0.0).iloc[-len(y):]
    Xj,Yj= make_supervised(pd.Series(y), lookback=lookback, horizon=1, extra_features=lag)
    return X,Y,Xj,Yj

from tensorflow.keras import callbacks
def metrics(y_true, y_pred):
    from sklearn.metrics import mean_absolute_error, mean_squared_error
    mae = float(mean_absolute_error(y_true, y_pred))
    rmse= float(np.sqrt(mean_squared_error(y_true, y_pred)))
    eps = 1e-8
    mape= float(np.mean(np.abs((y_true - y_pred)/np.maximum(np.abs(y_true), eps)))*100)
    return mae, rmse, mape

rows=[]
for ds, cfg in DATASETS.items():
    y = cfg["series"]()
    lookback, test_len = cfg["lookback"], cfg["test_len"]
    X,Y,Xj,Yj = make_XY_for(y, lookback)
    split = len(Y) - test_len
    for _, r in best_pm[best_pm["dataset"]==ds].iterrows():
        if r["model"]=="jordan":
            Xuse, Yuse = Xj, Yj
        elif r["model"]=="multi":
            Xuse, Yuse = X, Y
        else:
            Xuse, Yuse = X, Y
        Xtr, Ytr = Xuse[:split], Yuse[:split]
        Xte, Yte = Xuse[split:], Yuse[split:]

        m = build_from_row(r, input_shape=Xuse.shape[1:])
        es = callbacks.EarlyStopping(patience=8, restore_best_weights=True, monitor="val_loss")
        m.fit(Xtr, Ytr, validation_split=0.1, epochs=50, batch_size=64, verbose=0, callbacks=[es])
        pred = m.predict(Xte, verbose=0).ravel()
        MAE, RMSE, MAPE = metrics(Yte, pred)
        rows.append(dict(dataset=ds, model=r["model"], optimizer=r["optimizer"], lr=r["lr"],
                         units=r.get("units", np.nan), units1=r.get("units1", np.nan), units2=r.get("units2", np.nan),
                         test_MAE=MAE, test_RMSE=RMSE, test_MAPE=MAPE))
final = pd.DataFrame(rows).sort_values(["dataset","test_MAE"]).reset_index(drop=True)
final.to_csv("final_holdout_results.csv", index=False)
final
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import mean_absolute_error, mean_squared_error


DATASETS = {
    "climate": prep_climate_daily(),
    "air_quality": prep_air_quality_daily(),
    "airpassengers": prep_airpassengers_monthly(),
    "covid": prep_covid_daily(),
    "nifty": prep_nifty_returns()
}

LOOKBACKS = {"climate": 30, "air_quality": 30, "airpassengers": 12, "covid": 21, "nifty": 30}
TEST_SIZES = {"climate": 150, "air_quality": 150, "airpassengers": 18, "covid": 150, "nifty": 150}

results = []

for name, y in DATASETS.items():
    print(f"\n--- Dataset: {name} ---")
    test_len = TEST_SIZES[name]
    train, test = y[:-test_len], y[-test_len:]

    # ADF Test
    adf_p = adfuller(train)[1]
    d = 0 if adf_p < 0.05 else 1  # Difference once if not stationary

    # Try simple ARIMA orders (p, d, q) via AIC minimization
    best_aic = np.inf
    best_order = None
    best_model = None

    for p in range(4):
        for q in range(4):
            try:
                model = ARIMA(train, order=(p, d, q)).fit()
                if model.aic < best_aic:
                    best_aic = model.aic
                    best_order = (p, d, q)
                    best_model = model
            except:
                continue

    print(f"Best ARIMA order for {name}: {best_order}, AIC={best_aic:.2f}")

    forecast = best_model.forecast(steps=test_len)
    mae = mean_absolute_error(test, forecast)
    rmse = np.sqrt(mean_squared_error(test, forecast))
    mape = np.mean(np.abs((test - forecast) / np.maximum(np.abs(test), 1e-8))) * 100

    print(f"MAE: {mae:.4f}, RMSE: {rmse:.4f}, MAPE: {mape:.2f}%")

    results.append(dict(dataset=name, model="ARIMA", order=best_order, MAE=mae, RMSE=rmse, MAPE=mape))

# Save to compare later
arima_df = pd.DataFrame(results)
arima_df.to_csv("arima_results.csv", index=False)
arima_df


```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Parameters
n_folds = 4
train_start = 0
min_train = 20
val_size = 10
step = 10

# Plot
fig, ax = plt.subplots(figsize=(10, 2))

for i in range(n_folds):
    train_end = min_train + i * step
    val_start = train_end
    val_end = val_start + val_size

    # Draw training block
    ax.add_patch(patches.Rectangle((train_start, -i), train_end - train_start, 0.8,
                                   edgecolor='black', facecolor='lightgray', label='Train' if i == 0 else ""))

    # Draw validation block
    ax.add_patch(patches.Rectangle((val_start, -i), val_size, 0.8,
                                   edgecolor='black', facecolor='skyblue', label='Validation' if i == 0 else ""))

    # Label
    ax.text(val_end + 1, -i + 0.25, f"Fold {i+1}", va='center', fontsize=9)

# Formatting
ax.set_ylim(-n_folds, 1)
ax.set_xlim(0, val_end + 15)
ax.axis('off')
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.title("Walk-Forward (Expanding Window) Cross-Validation", fontsize=12)
plt.tight_layout()
plt.show()
```
